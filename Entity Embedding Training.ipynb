# 데이터 전처리

```python
# Nan 데이터 삭제
'''
data.dropna(axis=0, inplace=True)
data.shape
!이렇게 해버리면 데이터 개수가 424317개가 된다 -> 이러면 6으로 나누어 떨어지지 않아서 문제가 발생한다 
! 그렇기 때문에 for으로 해서 없애버리기 container id를 없애버리기
'''
print("결측치 데이터 삭제하기 전의 container id 개수 : ", len(data['Case ID'].unique()))
container_list = list(data['Case ID'].unique())

# train 데이터 생성
train_df = pd.DataFrame(columns=data.columns)

for container in tqdm(container_list, "결측치 제거중") :
    temp_df = data[data['Case ID'] == container].reset_index(drop=True)    
    num_nan = temp_df.isnull().sum().sum()
    # 만약에 결측치가 있다면, 그 container id는 train data에 포함시키지 않는다
    if  num_nan== 0 :
        train_df = pd.concat([train_df, temp_df])

        
```

LD를 2017년에 도착했던 선박에서 하는 경우가 있음

```python
# 데이터 프레임을 다시 학습할 수 있도록 데이터를 변환
# 생각하고 있는 Feature : 
#!   - Continous Data : DS 작업 3개의 작업 시간
#!   - Categorical Data(With Entity Embedding) : Container ID, VesselID(DS와 LD를 따로 생각), POD(LD의 POD)
#!   - Categorical Data(Without Entity Embedding) : FULL_EMPTY
# 생각하지 않은 Feature :
#!   - Resource : 이에 대한 정보는 Activity에 이미 담겨져 있음
#!   - MCHN_ID : 중요할까...? 이맘 논문에서는 어느정도 의미가 있었던 걸로 알지만 그거는 handling time 관점에서 
#!   - BLOCK_BAY : 일단 삭제 (중요해보임 - 토의 필요)

# 빈 데이터 프레임 생성(최종 사용할 데이터)
last_df = pd.DataFrame(columns=["Time of DS-QUAYSIDE", "Time of DS-MOVE", "Time of DS-YARDSIDE",
                                "ContainerID", "DS-VESSEL-ID", "LD-VESSEL-ID", "DS-VESSEL-YEAR", "LD-VESSEL-YEAR", "LD-POD", "Full", "Empty", "Dwell Time"])
i = 0
index = 0
# Continous Data 변환
for container in tqdm(container_list, desc = "학습할 데이터 생성 중") : 
    temp_df = data[data['Case ID'] == container].reset_index(drop=True)
    # DS-QUAYSIDE : 0 / DS-MOVE : 1 / DS-YARDSIDE : 2 / LD-YARDSIDE : 3 / LD-MOVE : 4 / LD-QUAYSIDE : 5
    # Time of DS-QUAYSIDE 계산
    start_DS_QUAYSIDE = temp_df[temp_df['Activity'] == 'DS-QUAYSIDE'].loc[:, 'Start Timestamp'][0]
    end_DS_QUAYSIDE = temp_df[temp_df['Activity'] == 'DS-QUAYSIDE'].loc[:, 'Complete Timestamp'][0]
    time_of_DS_QUAYSIDE = (end_DS_QUAYSIDE - start_DS_QUAYSIDE).seconds
    
    # Time of DS-MOVE 계산
    start_DS_MOVE = temp_df[temp_df['Activity'] == 'DS-MOVE'].loc[:, 'Start Timestamp'][1]
    end_DS_MOVE = temp_df[temp_df['Activity'] == 'DS-MOVE'].loc[:, 'Complete Timestamp'][1]
    time_of_DS_MOVE = (end_DS_MOVE - start_DS_MOVE).seconds
    
    # Time of DS-YARDSIDE 계산
    start_DS_YARDSIDE = temp_df[temp_df['Activity'] == 'DS-YARDSIDE'].loc[:, 'Start Timestamp'][2]
    end_DS_YARDSIDE = temp_df[temp_df['Activity'] == 'DS-YARDSIDE'].loc[:, 'Complete Timestamp'][2]
    time_of_DS_YARDSIDE = (end_DS_YARDSIDE - start_DS_YARDSIDE).seconds
    
    # Dwell Time(Target) 계산 
    end_DS = temp_df[temp_df['Activity'] == 'DS-YARDSIDE'].loc[:, 'Complete Timestamp'][2]
    start_LD = temp_df[temp_df['Activity'] == 'LD-YARDSIDE'].loc[:, 'Start Timestamp'][3]
    dwell_time = (start_LD - end_DS).days * 24 * 60 * 60 + (start_LD - end_DS).seconds
    
    # ContainerID
    # 혹시 6개의 작업의 Container 들이 다를 수 있으니 이를 확인
    if (temp_df.loc[0, "Case ID"] == temp_df.loc[1, "Case ID"] == temp_df.loc[2, "Case ID"] == temp_df.loc[3, "Case ID"]
        == temp_df.loc[4, "Case ID"] == temp_df.loc[5, "Case ID"]) :
        container_id = temp_df.loc[3, "Case ID"][:4]
    else :
        print(container, "의 경우 컨테이너 ID가 일치하지 않습니다")
        break

    # DS-Vessel ID
    # 혹시 3개의 작업에 Vessel 아이디가 다를 수도 있으니 이를 확인
    if (temp_df.loc[0, "VESSEL"].split("/")[0] == temp_df.loc[1, "VESSEL"].split("/")[0] == temp_df.loc[2, "VESSEL"].split("/")[0]) :
        ds_vessel_id = temp_df.loc[0, "VESSEL"][:4]
    else :
        print(container, "의 경우 3개의 DS-VESSEL ID가 일치하지 않습니다")
        break
    
    # LD-Vessel ID
    # 혹시 3개의 작업에 Vessel 아이디가 다를 수도 있으니 이를 확인
    if (temp_df.loc[3, "VESSEL"].split("/")[0] == temp_df.loc[4, "VESSEL"].split("/")[0] == temp_df.loc[5, "VESSEL"].split("/")[0]) :
        ld_vessel_id = temp_df.loc[4, "VESSEL"][:4]
    else :
        print(container, "의 경우 3개의 LD-VESSEL ID가 일치하지 않습니다")
        break

    # DS-Vessel Year
    # 혹시 3개의 작업에 Vessel Year가 다를 수도 있으니 이를 확인
    if (temp_df.loc[0, "VESSEL"].split("/")[1] == temp_df.loc[1, "VESSEL"].split("/")[1] == temp_df.loc[2, "VESSEL"].split("/")[1]) :
        ds_vessel_year = temp_df.loc[0, "VESSEL"].split("/")[1]
    else :
        print(container, "의 경우 3개의 DS-VESSEL 년도가 일치하지 않습니다")
        break
    
    # LD-Vessel Year
    # 혹시 3개의 작업에 Vessel 아이디가 다를 수도 있으니 이를 확인
    if (temp_df.loc[3, "VESSEL"].split("/")[1] == temp_df.loc[4, "VESSEL"].split("/")[1] == temp_df.loc[5, "VESSEL"].split("/")[1]) :
        ld_vessel_year = temp_df.loc[4, "VESSEL"].split("/")[1]
    else :
        print(container, "의 경우 3개의 LD-VESSEL 년도가 일치하지 않습니다")
        break
    
    # LD-POD
    # 점검
    if (temp_df.loc[3, "POD"] == temp_df.loc[4, "POD"] == temp_df.loc[5, "POD"]) :
        pod_ld = temp_df.loc[3, "POD"]
    else :
        print(container, "의 경우 3개의 LD의 POD가 일치하지 않습니다")
        break
    
    # FULL / EMPTY
    # 6가지의 경우에 다 같은지 확인
    if (temp_df.loc[0, "FULL_EMPTY"] == temp_df.loc[1, "FULL_EMPTY"] == temp_df.loc[2, "FULL_EMPTY"]
        == temp_df.loc[3, "FULL_EMPTY"] == temp_df.loc[4, "FULL_EMPTY"] == temp_df.loc[5, "FULL_EMPTY"]) :
        if (temp_df.loc[0, "FULL_EMPTY"] == "F") :
            full, empty = 1, 0
        elif (temp_df.loc[0, "FULL_EMPTY"] == "M") :
            full, empty = 0, 1
    else :
        print(container, "의 경우 6개의 FULL/EMPTY가 일치하지 않습니다")
        break
    
    row_list = [time_of_DS_QUAYSIDE, time_of_DS_MOVE, time_of_DS_YARDSIDE, container_id, ds_vessel_id, ld_vessel_id,
                ds_vessel_year, ld_vessel_year, pod_ld, full, empty, dwell_time]
    last_df.loc[i, :] = row_list
    i += 1
```
